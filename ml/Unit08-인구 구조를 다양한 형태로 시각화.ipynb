지역명을 입력받아서 연령별 인구수 데이터 시각화 하기

```python
import csv
f= open('age.csv', encoding = 'cp949')
data=csv.reader(f)

res = []

#지역명을 입력받는 변수 선언
name = input("인구 구조가 알고 싶은 지역의 이름( 읍면동 단위)을 입력해주세요 : ")

for row in data:
  if name in row[0]:
    for i in row[3:]:
      res.append(int(i.replace(',','')))  #인구수 데이터 정수로 변환 후 res 에 변환값 입력

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

plt.title(name+'지역의 인구 구조')
plt.plot(res)
plt.show()     
```

막대그래프 그리기 -> 비교 (지역별,성별 기준 등)

```python
import csv
import matplotlib.pyplot as plt

f= open('age.csv', encoding = 'cp949')
data=csv.reader(f)
print(next(data))

res = []

for row in data:
  if '새롬동' in row[0]:
    for i in row[3:]:
      res.append(int(i.replace(',',''))) #, 로 인해 형변환 오류 발생. replace로 제거

plt.figure(dpi=200)
plt.bar(range(101),res)
plt.show()
```

성별 데이터 저장하기 (type1)

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv', encoding = 'cp949')
data=csv.reader(f)
m =[]
f =[]

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)]))  # -1 값부터 담음 -> 100세 이상 부터 값이 담김

f.reverse()   # 0세부터 인구수 나오도록 뒤집기 

print(m)
print(f)
```

성별 데이터 저장하기 (type2)

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv', encoding = 'cp949')
data=csv.reader(f)
m =[]
f =[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]:    # 남성 : 3~103번 데이터 슬라이싱 
      m.append(int(i))
    for i in row[106:]:     # 여성 : 106~ 끝까지 데이터 슬라이싱
      f.append(int(i)) 

print(m)
print(f)
```

```python
plt.figure(dpi=200)
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv', encoding = 'cp949')
data=csv.reader(f)
m =[]
f =[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]:    # 남성 : 3~103번 데이터 슬라이싱 
      m.append(-int(i))
    for i in row[106:]:     # 여성 : 106~ 끝까지 데이터 슬라이싱
      f.append(int(i)) 

print(m)
print(f)

plt.figure(dpi=200)
plt.barh(range(101),m,label ='남성')
plt.barh(range(101),f,label='여성')
plt.legend()
plt.title("남성 여성 항아리")

plt.show()
```
