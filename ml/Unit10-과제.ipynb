꺾은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv', encoding = 'cp949')
data=csv.reader(f)
m =[]
f =[]

name = input("인구 구조가 알고 싶은 지역의 이름( 읍면동 단위)을 입력해주세요 : ")


for row in data:
  if name in row[0]:
    for i in row[3:104]:    # 남성 : 3~103번 데이터 슬라이싱 
      m.append(int(i.replace(',','')))
    for i in row[106:]:     # 여성 : 106~ 끝까지 데이터 슬라이싱
      f.append(int(i.replace(',',''))) 
    break

#꺾은선 그래프
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

plt.plot(m, label='남성')
plt.plot(f, label='여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv', encoding = 'cp949')
data=csv.reader(f)

name = input("찾고 싶은 지역의 이름( 읍면동 단위)을 입력해주세요 : ")

res = []

for row in data:
  if name in row[0]:
    for i in range(3,104):
      res.append(int(row[i].replace(',','')) -int(row[i+103].replace(',','')))
    break

 
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

plt.bar(range(101),res,color='r')
plt.title(name +" 지역의 남녀 성별 인구분포")

plt.show()
```

위치,크기가 서로 다른 100개의 점 만들기

```python
import random

x = []
y = []
size = []

for i in range(100):
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(10,100))

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y,s=size)
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y,s=size,c=size,cmap='jet')
plt.colorbar()
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha=0.8)# alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명

plt.colorbar()
plt.show()
```
