연령별 인구수(0~100세 이상)데이터 확인 

```python
import csv
f= open('age.csv', encoding = 'cp949')
data=csv.reader(f)
print(next(data))

for row in data:
  if '신도림' in row[0]:
    print(row[3:])   #0~100세 이상 인구수 출력 row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수
    print(len(row[3:]))  #0~100세 이상 데이터 수
```

우리 동네 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해

```python
import csv
f= open('age.csv', encoding = 'cp949')
data=csv.reader(f)
print(next(data))

res = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      res.append(int(i)) #인구수 데이터 정수로 변환 후 res 변수에 담기

print(res)   
```

우리 동네 연령별 인구수 데이터 시각화하기

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.style.use('dark_background')
plt.plot(res)
plt.show()
```
