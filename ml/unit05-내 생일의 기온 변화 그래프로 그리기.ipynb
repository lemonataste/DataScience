```python
#최고기온 데이터만 리스트에 담아서 출력하기

import csv

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '': # 최고 기온 데이터 값이 결측치가 아니라면
    result.append(float(row[-1]))

print(result) # 최고 기온 리스트 출력
```

```python
f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

res = [] #최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성
for row in data :
  if row[-1] != '':   #최고 기온 결측치 제외
    if row[0].split('-')[1] == '08': #날짜를 잘라서 08월 데이터만
      res.append(float(row[-1]))   #결과 리스트에 담기

plt.figure(dpi=300)
plt.plot(res,'deeppink') #res 리스트의 최고 기온 데이터를 그래프로 그리기
plt.show()
```

매년 생일 기준 최고 기온 그래프 그리기

```python
f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

res = [] #최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성
for row in data :
  if row[-1] != '':   #최고 기온 결측치 제외
    if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '22': #날짜를 잘라서 08월 데이터만
      res.append(float(row[-1]))   #결과 리스트에 담기

plt.figure(dpi=300)
plt.plot(res,'deeppink') #res 리스트의 최고 기온 데이터를 그래프로 그리기
plt.show()
```

매년 생일 기준 최고,최저 기온 그래프 그리기

```python
f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] 
low = []

for row in data :
  if row[-1] != '' and row[-2] != '':
    if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '22' :
      high.append(float(row[-1]))  #최고
      low.append(float(row[-2]))  #최저

plt.figure(dpi=300)
plt.plot(high,'deeppink', label = 'high')
plt.plot(low,'skyblue', label = 'low')  
plt.legend()
plt.show()
```

날짜 구간을 좁혀서 그래프 그리기

```python
f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] 
low = []

for row in data :
  if row[-1] != '' and row[-2] != '':
    if 1994 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] =='12' and row[0].split('-')[2] == '24':
        high.append(float(row[-1]))
        low.append(float(row[-2]))

#데이터 시각화
#차트 설정
plt.figure(dpi=200)
#차트 그리기
plt.plot(high,'deeppink', label = 'high')
plt.plot(low,'skyblue', label = 'low')
plt.title('내 생일의 기온 변화 그래프')  
plt.legend()
plt.show()
```

```python
f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = [] 
low = []
avg = []

for row in data :
  if row[-1] != '' and row[-2] != '' and row[-3 !='' ]: #최고 최저 평균 결측치 제거
    if 1994 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] =='12' and row[0].split('-')[2] == '24':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))

#데이터 시각화
#차트 설정
plt.figure(dpi=200)
#차트 그리기
plt.plot(high,'deeppink', label = 'high')
plt.plot(low,'skyblue', label = 'low')
plt.plot(avg,'g', label = 'avg')

plt.title('내 생일의 기온 변화 그래프')  
plt.legend()
plt.show()
```
