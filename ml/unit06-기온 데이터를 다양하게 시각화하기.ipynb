히스토그램 hist() 함수 사용

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200) #해상도
#히스토그램 -> 도수분포 -> 어떠한 자료에 속해있는 값들을 수량으로 나타낸 것
plt.hist([1,1,2,3,4,5,6,6,7,8,10]) #hist(data) 가로축은 계급(아이템), 세로축은 도수(횟수,빈도수)
plt.show()
```

주사위 시뮬레이션 - 히스토그램 연습

```python
import random as rd
import matplotlib.pyplot as plt

num = []  #주사위 값을 담을 변수

for a in range(5): 
  num.append(rd.randint(1,6)) #1~6 사이의 값을 랜덤으로 출력

#랜덤하게 굴린 5개의 값을 히스토그램으로 표현
plt.figure(dpi = 200)
plt.hist(num,bins=6)
plt.show()
```

최고 기온 데이터 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '': # 최고 기온 데이터 값이 결측치가 아니라면
    result.append(float(row[-1]))

#최고기온 데이터 히스토그램
plt.figure(dpi = 200)
plt.hist(result,bins=1000,color='r')
plt.show()
```

8월의 최고 기온 데이터 히스토그램으로 표현

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

aug = [] #8월의 최고 기온을 담을 변수

for i in data:
  month = i[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  day = i[0].split('-')[2]
  if i[-1] != '':
    if month == '02' and day =='22':
      aug.append(float(i[-1]))

plt.figure(dpi = 200)
plt.hist(aug,bins=100,color='r')
plt.show()
```

1월과 8월의 데이터를 히스토그램

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

aug = [] #8월의 최고 기온을 담을 변수
jan = [] #1월의 최고 기온을 담을 변수

for i in data:
  month = i[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  day = i[0].split('-')[2]

  if i[-1] != '':
    if month == '08' :
      aug.append(float(i[-1]))
    if month == '01' :
      jan.append(float(i[-1]))

plt.figure(dpi = 200)
plt.hist(aug,bins=100,color='r', label='aug')
plt.hist(jan,bins=100,color='b', label = 'jan')

plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

aug = [] #8월의 최고 기온을 담을 변수
jan = [] #1월의 최고 기온을 담을 변수
june =[] #6월의 최고 기온을 담을 변수

for i in data:
  month = i[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  day = i[0].split('-')[2]

  if i[-1] != '':
    if month == '08' :
      aug.append(float(i[-1]))
    if month == '01' :
      jan.append(float(i[-1]))
    if month == '04' :
      june.append(float(i[-1]))

plt.figure(dpi = 200)
plt.hist(aug,bins=100,color='r', label='aug')
plt.hist(jan,bins=100,color='b', label = 'jan')
plt.hist(june,bins=100,color='g', label = 'june')

plt.legend()
plt.show()
```

1월부터 12월까지 최고기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

month = [[],[],[],[],[],[],[],[],[],[],[],[]] #월별 데이터 저장할 변수

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

8월 1일부터 31일까지 최고기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

day = []

for i in range(31):
  day.append([]) #데이 리스트에 31개 리스트 생성

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(figsize=(10,5),dpi=200)
plt.boxplot(day,showfliers=False)
plt.show()
```
