출근 시간대 사람들이 가장 많이 타고 내리는 역은 어디일까

아침 7시 승차 데이터 추출 -> 10번 인덱스 데이터 값 추출


```python
import csv

f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

res = []  #아침 7시 승차데이터 추출 -> 10번째 인덱스에 존재

for row in data:
  row[4:] = map(int, row[4:])
  res.append(row[10])

print(len(res))  
```

아침 7시 승차 데이터 막대 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

res = []  #아침 7시 승차데이터 추출 -> 10번째 인덱스에 존재

for row in data:
  row[4:] = map(int, row[4:])
  res.append(row[10])

#데이터 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.bar(range(len(res)),res)
plt.show()
```

오름차순으로 데이터 정렬


```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

res = []  #아침 7시 승차데이터 추출 -> 10번째 인덱스에 존재

for row in data:
  row[4:] = map(int, row[4:])
  res.append(row[10])

res.sort()

#데이터 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.bar(range(len(res)),res)
plt.show()
```

7~9시 막대 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')

data = csv.reader(f)
next(data)
next(data)

result = [] #출근 시간(7시-9시) 승차 데이터 추출 -> 10, 12, 14번째 인덱스에 존재


for row in data:
  row[4:] = map(int,row[4:]) #리스트 값을 다 바꿔버린다 row[4:]를 int(정수)로 바꾼다
  result.append(sum(row[10 : 15 : 2])) #10 ,12, 14 추출

result.sort() #데이터를 오름차순으로 정렬

#데이터 시각화
plt.figure(dpi = 200)
plt.style.use('ggplot')
plt.bar(range(len(result)), result)
plt.show()
```

7시~9시까지 승차 인원 합계의 최댓값 찾기

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

mx=0
mx_st=''


for row in data:
  row[4:] = map(int, row[4:])
  if sum(row[10:15:2]) > mx:
    mx =  sum(row[10:15:2])
    mx_st = row[3] + '('+row[1]+')'

print(mx_st,mx)
```
