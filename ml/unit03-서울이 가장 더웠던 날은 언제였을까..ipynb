```python
import csv

max_temp = -999 #최고기온 값을 저장할 변수
max_date=''  # 최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f) #delimiter(구분자) -> ,로 생략가능
header = next(data) #첫번째 행을 읽은 후 다음 행으로 이동시키는 함수

for row in data:

  #최고기온 데이터가 '' (결측치) 확인
  if row[-1] =='':
    row[-1] = -999 #-999를 넣어서 빈 문자열(' ')이 있던 자리라고 표시

  row[-1] = float(row[-1])  #기온 데이터 실수형 변환

  #최고 기온 정보를 저장
  if max_temp < row[-1]:
    max_date = row[0]
    max_temp = row[-1]
f.close

print("기상 관측 이래 서울의 최고 기온이 가장 높았던 날은, ", max_date+'로',max_temp,'도 였습니다.')
```
